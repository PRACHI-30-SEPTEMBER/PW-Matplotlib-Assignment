## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

## Q2: What is a scatter plot? Use the following code to generate data and plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

**Answer:**
A scatter plot is used to display the relationship between two numerical variables. Each point represents a pair of values on the X and Y axis.

## Q3: Why is the subplot() function used? Draw four line plots using subplot().

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4,5])

y1 = np.array([0,100,200,300,400,500])
y2 = np.array([50,20,40,20,60,70])
y3 = np.array([10,20,30,40,50,60])
y4 = np.array([200,350,250,550,450,150])

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('Line 1')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Line 2')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('Line 3')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

**Answer:**
The subplot() function is used to create multiple plots in a single figure. It helps compare different datasets in one visualization.

## Q4: What is a bar plot? Why is it used? Plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(['Apple','Microsoft','Google','AMD'])
profit = np.array([3000,8000,1000,10000])

plt.bar(company, profit)
plt.title('Company Profit Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

**Answer:**
A bar plot is used to compare values between different categories. It represents data using rectangular bars.

## Q5: What is a box plot? Why is it used? Plot a box plot using the given data.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)

data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot Example')
plt.show()
```

**Answer:**
A box plot is used to show the distribution of data based on quartiles. It helps identify the median, spread, and outliers in a dataset.
